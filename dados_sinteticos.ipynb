Este bloco cria as tabelas e dados sintéticos para Clientes e Downloads

```python
import sqlite3
import random

#Conecta ao banco de dados SQLite
conn = sqlite3.connect("biblioteca.db")
cursor = conn.cursor()

#Cria tabela clientes
cursor.execute("""
    CREATE TABLE IF NOT EXISTS clientes (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        nome TEXT,
        sobrenome TEXT,
        cidade TEXT,
        estado TEXT,
        idade INTEGER CHECK(idade BETWEEN 18 AND 67)
    )
""")

# Criar tabela Downloads
cursor.execute("""
    CREATE TABLE IF NOT EXISTS downloads (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        cliente_id INTEGER,
        livro_id INTEGER,
        categoria INTEGER CHECK (categoria BETWEEN 1 AND 13),
        nota INTEGER CHECK (nota BETWEEN 1 AND 5),
        FOREIGN KEY (cliente_id) REFERENCES clientes(id)
    )
""")

conn.commit()
```
